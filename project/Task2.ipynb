```python
import numpy as np
def f(x): 
    return np.sin(x/5.0)*np.exp(x/10.0)+5*np.exp(-x/2.0)
```

```python
print(f(1), f(15))
```

```python
a = [[1, 1],[1, 15]]
b = [[f(1)], [f(15)]]
```

```python
import scipy
from scipy import linalg
data = linalg.solve(a,b)
data
```

```python
a = [[1, 1, 1],[1, 8, 64],[1, 15, 225]]
b = [f(1), f(8), f(15)]
data = linalg.solve(a,b)

rr = [1,8,15]
dat=[0,0,0]
dat[0] = data[0] + data[1]*rr[0] + data[2]*rr[0]*rr[0]
dat[1] = data[0] + data[1]*rr[1] + data[2]*rr[1]*rr[1]
dat[2] = data[0] + data[1]*rr[2] + data[2]*rr[2]*rr[2]

pyplot.plot(rr,dat)
dat
```
